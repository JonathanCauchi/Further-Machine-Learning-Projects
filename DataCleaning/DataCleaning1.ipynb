## Dataset obtained from Kaggle, link: https://www.kaggle.com/mansoordaku/ckdisease/downloads/ckdisease.zip/1

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('disease.csv')
```

```python
data.head()
```

```python
data.isnull() ## True indicates that a value is NaN
```

## How do we deal with missing data?

```python
cleaned = data.dropna()
```

```python
cleaned.head ## as you can see, all columns that included a NaN value have been dropped
```

```python
cleaned2 = data.dropna(how='all') ## this line drops all the rows that are all NaN
```

```python
cleaned2.head
```

#### We do not necessarily have to remove all NaN values, we can keep some columns or rows that consist some NaN values. This can be done via 'dropna(thresh=?)'

## Filling the Missing Data

#### Rather than removing rows and columns of data and possibly remove valuable information in the process, it would be ideal to 'fill' out the missing values. We can fill the missing values in a number of ways. The most popular one is the 'fillna' method

```python
fulld = data.fillna(0)
fulld.head() ## NaN values have been all replaced by 0
```

#### We can also specify the replacement value of each column as shown below

```python
fulld2 = data[['bp','sg']].fillna(value=1)
```

```python
fulld2.head ## as you can see the missing values have been replaced by 1
```

#### Removing values with a '?'

```python
data = data[(data.astype(str) != '?').all(axis=1)] ## add this line just in case
```

#### We can also pass the median or mean as the replacement value

```python
data = data.drop(['pcv','wc','rc'],axis=1) ## for some reason the NaN values of this column wont be replaced by the mean.
```

```python
data.fillna(data.mean(), inplace=True)
data.head
```

#### Checking for data duplicates

```python
data.duplicated() ## true being the row is duplicate
```

## Dealing with categorical variables that consist of string data

```python
data.head() 
```

```python
data.dropna(subset=['rbc','pc','pcc','ba','htn','dm','cad','appet','pe','ane','classification'], how='all', inplace = True)
## dropping rows of data that include NaN 
## problem with this is that we have a lot less data now
## ideally we have as much data as possible for model training
```

```python
dummies = pd.get_dummies(data=data, columns=['rbc','pc','ba','pcc','htn','dm','cad','appet','pe','ane','classification'], drop_first=True)
```

```python
dummies.isnull() ## no NaN values seem to be present, data seems to be tidy
```

## Lets save cleansed dataset into its own CSV file

```python
dummies.to_csv('diseaseclean.csv')
```
