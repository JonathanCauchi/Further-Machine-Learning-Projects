```python
import nltk
import matplotlib as plt
import pandas as pd
import numpy as np
from gensim.models import Word2Vec
# https://towardsdatascience.com/a-beginners-guide-to-word-embedding-with-gensim-word2vec-model-5970fa56cc92
```

```python
data = pd.read_csv('movie_review.csv')
```

```python
data.head(10)
```

```python
print(data.shape)
df = data[['text']]
print(df.head())
```

```python
df['text'][0]
```

```python
# Store them in a pandas dataframe
df_clean = pd.DataFrame({'clean': df['text']})
# Create the list of list format of the custom corpus for gensim modeling 
sent = [row.split(',') for row in df_clean['clean']]
```

```python
sentence = [nltk.word_tokenize(i) for i in df['text']]
```

```python
sentence[:20]
```

```python
model = Word2Vec(sentence, size=200, min_count=1, workers=4,sg=1)
```

```python
model['superman']
```

```python
print(model.similarity('superman', 'hero'))
print(model.similarity('superman', 'batman'))
print(model.similarity('superman', 'violence'))
print(model.similarity('superman', 'comics'))
```

```python
model.most_similar('superman')[:5]
```
