### The question we need to ask ourselves is, What economic and socioeconomic features should the ideal countries possess? Given our data, we can build a recommendation model by extracting the countries that correspond the best with these features:
##### - High GDP per capita
##### - Low unemployment rate
##### - High life expectancy at birth
##### - Low HIV/AIDS deaths

### Other features one may include could be low debt, recent GDP growth or  recent industrial growth but these dont really fit well. Greece for example is a country with a lot of debt but its hard to not consider it a holiday destination when you see photos of Santorini. So we will need to keep an open mind.

```python
recom_data = data[['Country','Death rate(deaths/1000 population)','HIV/AIDS - deaths','HIV/AIDS - people living with HIV/AIDS','HIV/AIDS - adult prevalence rate(%)','Life expectancy at birth(years)','GDP - per capita','Unemployment rate(%)','Internet users per capita']]
```

```python
recom_data.head(30)
```

```python
recom_data = recom_data.dropna(thresh=3) ## removed rows of data that have 3 or more NaN values
```

```python
recom_data.head(30)
```

```python
recom_data = recom_data[pd.notnull(recom_data['Death rate(deaths/1000 population)'])] ## removing countries that have no inforation on death rate
```

```python
recom_data.head(40)
```

### So when I think of poor countries, basically countries with a very low GDP per capita, I think of Mali for example. Given its such a poor country, there is a large probability that it is unsafe to travel. People there are probably living off scraps so foreigners might be preyed upon somehow. Another factor is sanitation, there is without a doubt a higher presence of diseases in these countries. So now lets classify the GDP per capita range that a 'poor' country falls upon (<= 1500) and we will print that list and see which countries pop up. We will analyze also other features of those countries and we will drop them from our list accordingly. L

```python
recom_data['GDP - per capita'] = recom_data['GDP - per capita'].astype(int)
poor_c =  recom_data['GDP - per capita']<=1500
data5 = recom_data[poor_c]
data5.head(25) ### these countries will be dropped
```

```python
recom_data = recom_data[recom_data['GDP - per capita'] > 1500]
```

```python
recom_data.shape
```

```python
recom_data.head(50)
```

### Next we will filter out countries who have a life expectancy less than that of 55 years

```python
recom_data['Life expectancy at birth(years)'] = recom_data['Life expectancy at birth(years)'].astype(float)
recom_data = recom_data[recom_data['Life expectancy at birth(years)'] >= 55]
recom_data.shape
```

### Next we will filter out countries with more than a 15% unemployment rate

```python
recom_data['Unemployment rate(%)'] = recom_data['Unemployment rate(%)'].astype(float)
recom_data = recom_data[recom_data['Unemployment rate(%)'] < 15]
recom_data.shape
```

```python
recom_data.head(20)
```

### Now we would like to view the countries that have a NaN value for 'HIV adult prevalence rate'. Some of these countries, albeit having a NaN values are quite decent and popular holiday destionations and shouldnt be removed from our recommendation list. So lets seperate them

```python
recom_data['HIV/AIDS - adult prevalence rate(%)'] = recom_data['HIV/AIDS - adult prevalence rate(%)'].astype(float)
null_data = recom_data[recom_data['HIV/AIDS - adult prevalence rate(%)'].isnull()]
null_data.shape

```

```python
null_data.head(40)
```

### Next we will filter out countries with an HIV prevalence rate of >=2%

```python
recom_data['HIV/AIDS - adult prevalence rate(%)'] = recom_data['HIV/AIDS - adult prevalence rate(%)'].astype(float)
k_data = recom_data[(recom_data['HIV/AIDS - adult prevalence rate(%)'] < 2) ] 
k_data.shape
```

```python
final_data = pd.concat([k_data, null_data])  ## MERGING THE 2 DATAFRAMES TOGETHER

final_data.shape
```

```python
final_data.head(50)
```
