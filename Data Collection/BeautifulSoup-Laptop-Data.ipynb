 The goal of the data extraction was to build a data driven laptop recommender chatbot using NLP techniques with the data extracted. Considering we dont have a lot of data on this site, we will use a seperate dataset obtained from Kaggle for that project. The purpose of this project is to show how one can manually extract data from a particular site so that you can build your own data driven applications :)

```python
url = "https://www.pcmag.com/roundup/256052/the-best-budget-laptops"
headers = {'User-Agent':'Mozilla/5.0'}
page = requests.get(url)
soup = BeautifulSoup(page.text, "html.parser")
```

Lets set the data into a nice CSV file

```python
import csv

f = open("laptopdata.csv", "w")
writer = csv.writer(f)
writer.writerow(["Laptop name", "Suitability", "Processor", "Platform", "Processor Speed", "RAM", "Storage", "Screen Size", "Resolution", "Weight", "Battery Duration", "Price"]) # Write column headers as the first line
```

```python
for i in range(len(objects)): 
    objects = soup.findAll("b",{"class":"review-title"})
    name = objects[i].text
    objects2 = soup.find("tr",{"class":"specs first"}).findAll("td",{"class":"secondcolumn"})
    suitability = objects2[i].text 
    objects3 = soup.find("tr",{"class":"specs odd"}).findAll("td",{"class":"secondcolumn"})
    processor = objects3[i].text
    objects4 = soup.find("tr",{"class":"specs even"}).findAll("td",{"class":"secondcolumn"})
    platform = objects4[i].text
    objects5 = soup.find("tr",{"class":"specs even","data-name":"Processor Speed"}).findAll("td",{"class":"secondcolumn"})
    procspeed = objects5[i].text
    objects6 = soup.find("tr",{"class":"specs odd","data-name":"RAM (as Tested)"}).findAll("td",{"class":"secondcolumn"})
    ram = objects6[i].text
    objects7 = soup.find("tr",{"class":"specs even","data-name":"Storage Capacity (as Tested)"}).findAll("td",{"class":"secondcolumn"})
    storage = objects7[i].text
    objects8 = soup.find("tr",{"class":"specs odd","data-name":"Screen Size"}).findAll("td",{"class":"secondcolumn"})
    screen = objects8[i].text
    objects9 = soup.find("tr",{"class":"specs even","data-name":"Resolution"}).findAll("td",{"class":"secondcolumn"})
    resolution = objects9[i].text
    objects10 = soup.find("tr",{"class":"specs odd","data-name":"Weight"}).findAll("td",{"class":"secondcolumn"})
    weight = objects10[i].text
    objects11 = soup.find("tr",{"class":"specs even","data-name":"Battery Rundown"}).findAll("td",{"class":"secondcolumn"})
    battery = objects11[i].text
    objects12 = soup.find("tr",{"class":"specs","data-name":"Msrp"}).findAll("td",{"class":"secondcolumn"})
    price = objects12[i].text
    
    writer.writerow([name, suitability, processor, platform, procspeed, ram, storage, screen, resolution, weight, battery, price])
    
    print("Name: " + name  )
    print("Suitabilty: " + suitability   )
    ### and so on
    
f.close()
    
    
```
