```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv("weight-height.csv")
```

```python
data.head
```

```python
data.describe()
```

```python
height = data['Height']
height_dist = sns.distplot(height)
height_dist.set_title("Distribution of Height amongst the population")
```

```python
data['Height'].hist(bins=65)
```

```python
sns.catplot('Gender', data=data, kind='count', aspect=1.5) ### use catplot instead of factorplot to avoid warning message
```

```python
fig = sns.FacetGrid(data, hue='Gender', size=3, aspect=5)
fig.map(sns.kdeplot, 'Weight', shade=True)
oldest = data['Weight'].max()
fig.set(xlim=(0,oldest+40))
fig.set(title='Distribution of Weight Grouped by Gender')
fig.add_legend()
```

```python
fig = sns.FacetGrid(data, hue='Gender', size=3, aspect=5)
fig.map(sns.kdeplot, 'Height', shade=True)
oldest = data['Height'].max()
newest = data['Height'].min()
fig.set(xlim=(newest-20,oldest+20))
fig.set(title='Distribution of Height Grouped by Gender')
fig.add_legend()
```

```python
sns.jointplot(y='Height',x='Weight',data=data)
```

```python
sns.scatterplot(x="Weight", y="Height", hue="Gender", data=data)
sns.set(rc={'figure.figsize':(11.7,8.27)})
```

```python
sns.boxplot(x='Gender', y='Weight', data=data)
```

```python
sns.boxplot(x='Gender', y='Height', data=data)
```

```python
ax = sns.violinplot(x="Gender", y="Weight", data=data)
```
