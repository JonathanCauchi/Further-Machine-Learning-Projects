```python
#import libraries
import random
```

```python
def random_walk(n):
    #init coordinates
    x,y = 0,0
   
    for i in range(n):
        step = random.choice(['N','S','W','E'])
        if step==' N':
            y += 1
        elif step == 'S':
            y-=1
        elif step =='W':
            x-=1
        else:
            x+=1
    return(x,y)

for i in range(25): # testing 25 times
    walks = random_walk(12) # we will be taking 12 random steps
    print(walks)
    print('Distance from initial poisition: ', walks[0] + walks[1])
    
```

### Lets say that, for random walks that lead us astray more than a value of 4, we will have to take a taxi back home. So given our history of random walks, what is the probability of us needing to take a taxi after a random walk of 12 steps. Lets compute this

```python
count = 0
n=25
for i in range(n): # testing 25 times
    walks = random_walk(12) # we will be taking 12 random steps
    if( (walks[0] + walks[1])  > 4):
        count += 1
    elif ( (walks[0] + walks[1])  < -4):
        count += 1
    print(walks)
    print('Distance from initial poisition: ', walks[0] + walks[1])
    
print('\n')
print(count)
print('\n')
print('Probability of taking the taxi on the next random walk is:', ((count)/n)*100 ,'%')
    
```
